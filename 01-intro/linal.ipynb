```python
u = np.array([1, 4, 3, 5])
v = np.array([3, 8, 9, 6])

def vector_vector_multiplication(a, b):
    assert a.shape[0] == b.shape[0] 
    n = a.shape[0]
    result = 0.0
    for i in range(n):
        result +=a[i]*b[i]
    return result

vector_vector_multiplication(u, v)
```

```python
u.dot(v)
```

```python
U = np.array([[2, 3, 4, 5], [8, 7, 5, 6], [9, 2, 8, 3]])

def matrix_vector_multiplication(A, b):
    assert A.shape[1] == b.shape[0] 
    n = A.shape[0]
    result = np.zeros(n)
    for i in range(n):
        result[i] += vector_vector_multiplication(A[i], b)
    return result

matrix_vector_multiplication(U, v)
```

```python
U.dot(v)
```

```python
V = np.array([[2, 3, 4], [8, 7, 5], [9, 2, 8], [4, 6, 7]])

def matrix_matrix_multiplication(A, B):
    assert A.shape[1] == B.shape[0] 
    n = A.shape[0]
    m = B.shape[1]
    result = np.zeros([n, m])
    for i in range(n):
        for j in range(m):
            result[i][j] += vector_vector_multiplication(A[i], B[:, j])
    return result

matrix_matrix_multiplication(U, V)
```

```python
U.dot(V)
```
